```python
import seaborn as sns
import matplotlib.pyplot as plt

# Before filtering
plt.figure(figsize=(8, 4))
sns.boxplot(x=df["Area"])
plt.title("Before Removing Outliers")
plt.show()

# Outlier removal
low_outliers = df["Area"].quantile(0.1)
high_outliers  = df["Area"].quantile(0.90)
df = df[(df["Area"] < high_outliers) & (df["Area"] > low_outliers)]
df = df.reset_index(drop=True)

# After filtering
plt.figure(figsize=(8, 4))
sns.boxplot(x=df["Area"])
plt.title("After Removing Outliers")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

ax = df["State"].value_counts().plot.bar(
    figsize=(18, 8),
    color=cm.get_cmap('Spectral')(np.linspace(0, 1, df["State"].nunique()))
)
plt.title("Crop Records by State")
plt.xlabel("State")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

season_counts = df["Season"].value_counts(ascending=True)
ax = season_counts.plot.barh(
    figsize=(18, 8),
    color=cm.get_cmap('plasma')(np.linspace(0, 1, len(season_counts)))
)
plt.title("Crop Records by Season")
plt.xlabel("Count")
plt.ylabel("Season")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

crop_counts = df["Crop"].value_counts()
ax = crop_counts.plot.bar(
    figsize=(22, 8),
    color=cm.get_cmap('viridis')(np.linspace(0, 1, len(crop_counts)))
)
plt.title("Crop Records by Crop Type")
plt.xlabel("Crop")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```
